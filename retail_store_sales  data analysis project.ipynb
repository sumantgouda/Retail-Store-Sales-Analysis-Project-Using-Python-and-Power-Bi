```python
df["Price Per Unit"]=df["Price Per Unit"].interpolate(method="linear")
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
df["Quantity"]=df["Quantity"].fillna(df["Quantity"].mean())
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
df["Quantity"]=df["Quantity"].round()
```

```python
df.head(20)
```

```python
df["Total Spent"]=df["Price Per Unit"]*df["Quantity"]
```

```python
df["Total Spent"].isnull().sum()
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df["Discount Applied"]=df["Discount Applied"].ffill()
```

```python
df.head(20)
```

```python
df["Item"]=df["Item"].ffill()
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df["Item"]=df["Item"].str.split("_").str[-1]
```

```python
df.head(20)
```

```python
category_item_map = (
    df.dropna(subset=["Item"])
      .groupby("Category")["Item"]
      .agg(lambda x: x.mode()[0])
)
```

```python
df["Item"] = df["Item"].fillna(df["Category"].map(category_item_map))
```

```python
df.isnull().sum()
```
